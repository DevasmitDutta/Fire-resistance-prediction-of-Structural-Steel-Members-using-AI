```python
#build model

# compile model
adam = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer = adam, loss = 'mean_squared_error',metrics=['mae'])
epochs_hist = model.fit(X_train, y_train, epochs = 48000, batch_size = 50, validation_split = 0.2)
model.save('Steel-columns ANN using DNN .h5')

#evaluate model
plt.plot(epochs_hist.history['loss'])
plt.plot(epochs_hist.history['val_loss'])
plt.legend(['Training loss', 'Validation loss'])
plt.show()
```

```python
plt.plot(epochs_hist.history['mean_absolute_error'])
plt.plot(epochs_hist.history['val_mean_absolute_error'])
plt.legend(['Training mae', 'Validation mae'])
plt.show()
```

```python
y_pred_1 = model.predict(X_test)
y_pred_1 = scaler.inverse_transform(y_pred_1)
y_test = scaler.inverse_transform(y_test)
plt.plot(y_test,y_pred_1,'ob',y_test,y_test,'--k')
plt.xlabel('Actual')
plt.ylabel('Predicted')
#plt.plot(y_test,y_test,'--k')
```

```python
import math
from sklearn import metrics 
mape  = metrics.mean_absolute_percentage_error(y_pred_1,y_test)
mae = metrics.mean_absolute_error(y_pred_1, y_test)
mse = metrics.mean_squared_error(y_pred_1, y_test)
r2 = metrics.r2_score(y_pred_1, y_test)

print("The model performance for testing set")
print("--------------------------------------")
print("MAPE is {}",format(mape))
print('MAE is {}'.format(mae))
print('RMSE is {}'.format(math.sqrt(mse)))
print('R2 score is {}'.format(r2))
```
